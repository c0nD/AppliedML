```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Creating a 2x2 grid of plots

fig, axes = plt.subplots(2,2)
ax1_1, ax1_2, ax2_1, ax2_2 = axes.ravel()

plt.tight_layout()
```

```python
# Another way to plot a 2x2 grid

ax11 = plt.subplot2grid((2,2), (0,0))
ax12 = plt.subplot2grid((2,2), (0,1))
ax21 = plt.subplot2grid((2,2), (1,0))
ax22 = plt.subplot2grid((2,2), (1,1))

x = np.linspace(0, 2*np.pi, 400)
y = np.sin(x**2)

ax11.plot(x, y)
```
