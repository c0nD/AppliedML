### #1 
Generate a scatter plot of longitude against latitude using the default scatter plot of
Matplotlib.

```python
plt.figure(figsize=(7, 5))
plt.scatter(df['longitude'], df['latitude'])
plt.title('Basic Scatter Plot of Longitude vs Latitude')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.show()
```

### #2
Based on the first axes, fine-tune the alpha and marker size parameters to address any
issues of overplotting.

```python
plt.figure(figsize=(7, 5))
plt.scatter(df['longitude'], df['latitude'], alpha=0.2, s=10)
plt.title('Scatter Plot with Alpha and Marker Size Tuning')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.show()
```

### #3
Based on the second axes, employ varying colors to depict Brightness Temperature I-4
(bright_ti4) values.

```python
plt.figure(figsize=(7, 5))
sc = plt.scatter(df['longitude'], df['latitude'], c=df['bright_ti4'], cmap='viridis', alpha=0.5, s=10)
plt.colorbar(sc, label='Brightness Temperature I-4')
plt.title('Scatter Plot with Brightness Temperature I-4 Coloring')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.show()
```

### #4
Illustrate the distribution of Brightness Temperature I-4 (bright_ti4) by utilizing a
histogram. Modify the density setting to highlight potential gaps within the distribution.


```python
plt.figure(figsize=(7, 5))
plt.hist(df['bright_ti4'], bins=50, density=True, alpha=0.7, color='g')
plt.title('Distribution of Brightness Temperature I-4')
plt.xlabel('Brightness Temperature I-4')
plt.ylabel('Density')
plt.show()
```
